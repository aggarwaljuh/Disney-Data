```python
import pandas as pd
import numpy as np

credit_data = "data/credits.csv"
title_data = "data/titles.csv"

credit_df = pd.read_csv(credit_data)
title_df = pd.read_csv(title_data)
```

```python
title_df.head()
```

```python
#only shows
titleTV_df=title_df.loc[title_df['type']=="SHOW"]

titleTV_df.head()
```

```python
#remove NAN
#Clean data

titleTV_df.dropna()
```

```python
by_year_ave=pd.DataFrame(titleTV_df.groupby(['release_year'])['imdb_score'].mean())
by_year_ave.reset_index(inplace=True)
year_count=pd.DataFrame(titleTV_df.groupby(['release_year'])['imdb_score'].count())
year_count.reset_index(inplace=True)
year_merged = pd.merge(by_year_ave,year_count,how='inner', on = 'release_year')
year_merged=year_merged.rename(columns={"release_year": "Year", "imdb_score_x": "Average", "imdb_score_y": "Count"})
year_merged
```

```python
ax1 = year_merged.plot.scatter(x='Year', y='Average',c="Count",colormap='viridis')
```
